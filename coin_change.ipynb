```python
#Greedy approach values for the normal values

import pandas as pd
import time
import matplotlib.pyplot as plt

def coin_change_greedy_with_time(S, denominations):
    results = []
    denominations.sort(reverse=True)

    for k in range(1, len(denominations) + 1):
        selected_denominations = denominations[:k]
        remaining = S
        exact_solution = True

        start_time = time.time()

        coin_counts = {}

        for coin in selected_denominations:
            count = remaining // coin
            if count > 0:
                coin_counts[coin] = int(count)
                remaining = round(remaining - coin * count, 2)
            else:
                coin_counts[coin] = 0

        end_time = time.time()
        time_taken = end_time - start_time

        if remaining > 0:
            exact_solution = False
            print(f"Warning: Remaining amount {remaining} could not be changed with denominations {selected_denominations}")

        results.append({
            "k": k,
            "Denominations Used": selected_denominations,
            "Coins Used": coin_counts,
            "Total Coins Used": sum(coin_counts.values()),
            "Time (seconds)": time_taken,
            "Exact Solution": exact_solution
        })

    return results

S = 125.5
denominations = [2.0, 1.0, 0.6, 0.5, 0.2, 0.1, 0.05, 0.03, 0.02, 0.01]

results = coin_change_greedy_with_time(S, denominations)
df_results = pd.DataFrame(results)
df_results['Coins Used'] = df_results['Coins Used'].apply(lambda coins: ', '.join([f"{k}€: {v}" for k, v in coins.items()]))
print(df_results.to_string(index=False))

fig, ax1 = plt.subplots()
ax1.set_xlabel('Number of Denominations (k)')
ax1.set_ylabel('Time Taken (seconds)', color='tab:red')
ax1.plot(df_results['k'], df_results['Time (seconds)'], color='tab:red', marker='o', label='Time Taken')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Coins Used', color='tab:blue')
ax2.plot(df_results['k'], df_results['Total Coins Used'], color='tab:blue', marker='s', label='Total Coins Used')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Runtime and Total Coins Used for Coin Change Greedy Algorithm')
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
from collections import defaultdict
import random

def coin_change_dp_all_combinations(S, denominations):
    results = []

    # Shuffle the denominations to randomize the order
    randomized_denominations = denominations[:]
    random.shuffle(randomized_denominations)

    for k in range(1, len(randomized_denominations) + 1):
        selected_denominations = [int(coin * 100) for coin in randomized_denominations[:k]]
        S_cents = int(S * 100)

        start_time = time.time()
        dp = [set() for _ in range(S_cents + 1)]
        dp[0].add(tuple())

        for coin in selected_denominations:
            for amount in range(coin, S_cents + 1):
                for combination in dp[amount - coin]:
                    new_combination = combination + (coin,)
                    dp[amount].add(new_combination)

        end_time = time.time()
        time_taken = end_time - start_time

        all_combinations = []
        for combination in dp[S_cents]:
            coin_counts = defaultdict(int)
            for coin in combination:
                coin_euro = coin / 100
                coin_counts[coin_euro] += 1
            all_combinations.append(dict(coin_counts))

        results.append({
            "k": k,
            "Denominations Used": [coin / 100 for coin in selected_denominations],
            "All Possible Combinations": all_combinations,
            "Total Combinations": len(all_combinations),
            "Time (seconds)": time_taken,
            "Exact Solution": len(all_combinations) > 0
        })

    return results

# Original set of denominations
denominations = [2.0, 1.0, 0.5, 0.2, 0.1, 0.05]
S = 12.85

results = coin_change_dp_all_combinations(S, denominations)
df_results = pd.DataFrame(results)
df_results['All Possible Combinations'] = df_results['All Possible Combinations'].apply(
    lambda combs: '\n'.join([', '.join([f"{coin}€: {count}" for coin, count in comb.items()]) for comb in combs[:5]])
    + ('\n...' if len(combs) > 5 else '')
)
print(df_results.to_string(index=False))

# Plotting results
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Number of Denominations (k)')
ax1.set_ylabel('Time Taken (seconds)', color='tab:red')
ax1.plot(df_results['k'], df_results['Time (seconds)'], color='tab:red', marker='o', label='Time Taken')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Total Combinations', color='tab:blue')
ax2.plot(df_results['k'], df_results['Total Combinations'], color='tab:blue', marker='s', label='Total Combinations')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Runtime and Total Combinations for Coin Change DP Algorithm with Randomized Denominations')
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

edges = {
    ('A', 'B'): 0.45, ('B', 'A'): 1.1,
    ('B', 'C'): 1, ('C', 'B'): 1,
    ('C', 'D'): 10, ('D', 'C'): 10,
    ('D', 'E'): 15, ('E', 'D'): 15
}

locations = ['A', 'B', 'C', 'D', 'E']
location_names = {
    'A': 'MM23 Room',
    'B': 'Hall',
    'C': 'Bus Stop',
    'D': 'Acrople',
    'E': 'Belfort Station'
}

n = len(locations)
adj_matrix = np.full((n, n), np.inf)
np.fill_diagonal(adj_matrix, 0)
location_index = {loc: idx for idx, loc in enumerate(locations)}
for (start, end), weight in edges.items():
    adj_matrix[location_index[start]][location_index[end]] = weight
print("Initial Matrix (Direct Distances):")
print(adj_matrix)

G = nx.DiGraph()
G.add_nodes_from(locations)

for (start, end), weight in edges.items():
    G.add_edge(start, end, weight=weight)


pos = nx.circular_layout(G)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, labels=location_names, node_size=2000, node_color='lightgreen', font_size=10, font_weight='bold')
edge_labels = {(u, v): f"{d['weight']}s" for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
plt.title("Initial Graph with Weights (Circular Layout)")
plt.show()

for k in range(n):
    for i in range(n):
        for j in range(n):
            adj_matrix[i][j] = min(adj_matrix[i][j], adj_matrix[i][k] + adj_matrix[k][j])

print("\nFinal Matrix after Transitive Closure (Shortest Path Distances):")
print(adj_matrix)

start_node = 'A'
end_node = 'E'
length, path = nx.single_source_dijkstra(G, source=start_node, target=end_node, weight='weight')
print(f"\nShortest path from {location_names[start_node]} to {location_names[end_node]}:")
print(" -> ".join([location_names[node] for node in path]))
print(f"Total travel time: {length} seconds")


plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, labels=location_names, node_size=2000, node_color='lightgreen', font_size=10, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
path_edges = list(zip(path, path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)
plt.title("Graph Showing Shortest Path from A to E (Circular Layout)")
plt.show()
```
